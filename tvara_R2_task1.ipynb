```python
from PyPDF2 import PdfReader
```

```python
def read_pdf(file_path):
  read = PdfReader(file_path)
  text = ''
  for page in read.pages:
    text += page.extract_text()
  return text.strip()
```

```python
#moderation check
def moderation_check(text):
  bann_word = ['kill', 'bad', 'hate', 'destroy']
  # bann_word = ['kill', 'hate', 'destroy']
  for word in bann_word:
    if word in text.lower():
      return{"allowed": False, "flag": word}
  return{"allowed": True}
```

```python
import time
```

```python
def retry(func, arg, retries=2, delay=1):
  for i in range(retries):
    try:
      return func(arg)
    except Exception as e:
      if i ==  retries-1:
        return{"error": str(e)}
      time.sleep(delay)
```

```python
#main logic
def pipeline(file_path):
  text = read_pdf(file_path)
  mod_result = moderation_check(text)
  if not mod_result["allowed"]:
    return mod_result

  response = retry(gpt_resp, text)

  return {"allowed": True, "response": response}
```
